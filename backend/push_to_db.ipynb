```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('A:\\Projects\\project_analytics\\Project Management Dataset.csv')
print(df.head())
```

```python
print("Original columns:")
print(df.columns.to_list())
```

```python
df.columns = (
    df.columns.str.strip()          # remove leading/trailing spaces
             .str.lower()           # lowercase
             .str.replace(" ", "_") # replace spaces with underscores
             .str.replace("%", "percent") # replace '%' with word
             .str.replace("/", "_") # replace '/' with underscore if any
)
```

```python
for col in ["project_cost", "project_benefit"]:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)

# 4️⃣ Clean completion percent (remove '%' and convert to float)
df["completionpercent"] = df["completionpercent"].replace('%', '', regex=True).astype(float)

# 5️⃣ Convert dates to datetime
df["start_date"] = pd.to_datetime(df["start_date"], errors='coerce')
df["end_date"] = pd.to_datetime(df["end_date"], errors='coerce')
```

```python
print("\nCleaned columns:")
print(df.columns.tolist())
```

```python
engine = create_engine("postgresql+psycopg2://postgres:}Brindavanam6@localhost:5432/project_analytics")
```

```python
df.to_sql("projects", engine, if_exists="replace", index=False)
print("Data inserted successfully")
print(pd.read_sql("SELECT * FROM projects LIMIT 5;", engine))
```
